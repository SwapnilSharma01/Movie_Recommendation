## Task 3: Data Analysis

-  ### Subtask 3.1: Convert units of gross and budget to millions for all movies in the dataset. (1 mark)

```python
# movies['gross'] = movies['gross']/1000000 will divide the gross column by million and will assign back to movies['gross'] column
movies['gross'] = movies['gross']/1000000
movies['gross']

# movies['budget'] = movies['budget']/1000000 will divide the gross column by million and will assign back to movies['budget'] column
movies['budget'] = movies['budget']/1000000
movies['budget']
```

-  ### Subtask 3.2:  Display the 10 movies with the highest profit. (Blockbuster Movies playlist) (1 mark)

```python

#1) movies['profit'] = movies['gross'] - movies['budget'] : This code will calculate the proft i.e. gross - budget and assign that to new profit column

#2) sort_values('profit', ascending = False).head(10) : Below code is sorting the gross column in descending order. 

#3) head(10) : displays the top 10 records

#4) movies.sort_values('profit', ascending = False).head(10) : Will displayDataFrame movies as output


movies['profit'] = movies['gross'] - movies['budget']
movies.sort_values('profit', ascending = False).head(10)
```

-  ### Subtask 3.3: Display the top 10 movies (based on IMDb ratings) with a minimum of 25,000 ratings (Most Loved movies playlist) (2 marks)

```python
# movies[ movies['num_voted_users'] >= 25000][['movie_title','imdb_score','num_voted_users']] : movies['num_voted_users'] >= 25000 will give records only where num_voted_users is equal to or greater than 25000.
#Then [['movie_title','imdb_score','num_voted_users']] will provide 'movie_title','imdb_score','num_voted_users' from the DataFrame movies.

# .sort_values('imdb_score', ascending = False).head(10) : This will sort the records on imdb_score in decending order and head will display only top 10 records.

# movies[ movies['num_voted_users'] >= 25000][['movie_title','imdb_score','num_voted_users']].sort_values('imdb_score', ascending = False).head(10) :
#This entire piece of code will provide top 10 records based on imdb_score where minimum ratings are 25000.

movies[ movies['num_voted_users'] >= 25000][['movie_title','imdb_score','num_voted_users']].sort_values('imdb_score', ascending = False).head(10)
```

- ### Subtask 3.4: Display the names of  the top 5 directors based on IMDb ratings with a minimum of 5 movies (Top Directors Playlist)  (2 marks)

```python

#1) df_temp1 = pd.DataFrame(movies.groupby('director_name').count()[['movie_title']]) :
#This code will provide the count of movies each director dircted and assign that group and count to DataFrame df_temp1

#2) df_temp1.rename(columns = {'movie_title':'directed_movie_count'}, inplace = True) :
#This code renames the column in df_temp1 where counts are saved to directed_movie_count. The rename is needed to provide correct representation of column

#3) df_temp2 = pd.DataFrame(df_temp1[df_temp1['directed_movie_count'] >= 5].sort_values('directed_movie_count' , ascending = False)) :
#The above code will find out the director list who have directed 5 or more than 5 movies and stores that data into new dataFrame df_temp2. 
#The code first find the directors directed 5 or more than 5 movies by df_temp1['directed_movie_count'] >= 5 and then sorts the filtered data sort_values('directed_movie_count' , ascending = False)) 
#df_temp2 stores the list of directors and count of directed movies for the directors who have directed 5 or more than 5 movies.   

#4) movies[movies['director_name'].isin(df_temp2.index)].groupby('director_name').mean()[['imdb_score']].sort_values(by = 'imdb_score', ascending = False).head(5) :
#  4a) movies['director_name'].isin(df_temp2.index) : From movies data frame, this code finds the rcords where director_name in df_temp2 DataFrame
#  4b) groupby('director_name').mean()[['imdb_score']] : This code groups the data from point 4a on director_name wtaking the mean on imdb_score
#  4c) sort_values(by = 'imdb_score', ascending = False).head(5) : This code sorts the data grouped from point 4b on imdb_score in descending order nad takes the top 5 records



df_temp1 = pd.DataFrame(movies.groupby('director_name').count()[['movie_title']])
df_temp1.rename(columns = {'movie_title':'directed_movie_count'}, inplace = True)
df_temp2 = pd.DataFrame(df_temp1[df_temp1['directed_movie_count'] >= 5].sort_values('directed_movie_count' , ascending = False))

movies[movies['director_name'].isin(df_temp2.index)].groupby('director_name').mean()[['imdb_score']].sort_values(by = 'imdb_score', ascending = False).head(5)
```

-  ### Subtask 3.5.1: Find the most popular genre of movies (by highest average rating) (2 marks)

```python
# movies['primary_genres'] = movies['genres'].str.split('|', expand=True)[0] : 
#    For the purpose of finding Primary genre, this code will split the data of genres column using str.split on '|' delimiter and takes the first set using [0] and assigns the same to a new column primary_genres in the movies DataFrame

#2) movies.groupby('primary_genres').mean()[['imdb_score']].sort_values(by = 'imdb_score',ascending = False).head(1) :
#    This code groups the primary_genres for mean of imdb_score. The data is then soted on imdb_score in descending order and top record is thus displayed.

movies['primary_genres'] = movies['genres'].str.split('|', expand=True)[0]

movies.groupby('primary_genres').mean()[['imdb_score']].sort_values(by = 'imdb_score',ascending = False).head(1)
```

-  ### Subtask 3.5.2: Find the most popular genre of movies (by total gross) (2 marks)

```python
#1) movies['primary_genres'] = movies['genres'].str.split('|', expand=True)[0] : 
#    For the purpose of finding Primary genre, this code will split the data of genres column using str.split on '|' delimiter and takes the first set using [0] and assigns the same to a new column primary_genres in the movies DataFrame

#2) movies.groupby('primary_genres').sum()[['gross']].sort_values(by = 'gross',ascending = False).head(1)
#    This code groups the primary_genres for sum of gross. The data is then soted on gross in descending order and top record is thus displayed.


movies['primary_genres'] = movies['genres'].str.split('|', expand=True)[0]

movies.groupby('primary_genres').sum()[['gross']].sort_values(by = 'gross',ascending = False).head(1)

```
